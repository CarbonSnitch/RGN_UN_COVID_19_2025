```python
# Create a correlation matrix using pandas
corr = df_UN_corr.corr()
```

# Create a correlation matrix heatmap (colored).
## Discuss what the coefficients in the plot mean in terms of the relationships between the variables.

```python
# Create figure first
fig_corr = plt.figure(figsize=(8, 8))  # figure size 

# Get the axes for this figure
ax = fig_corr.add_subplot(111)  # Create axes object

# Create the heatmap on this figure
im = ax.imshow(corr)  # type of plot, use ax instead of plt

# Add labels and formatting
ax.set_xticks(range(df_UN_corr.shape[1]))
ax.set_xticklabels(df_UN_corr.columns, fontsize=8, rotation=60)  # x axis labels
ax.set_yticks(range(df_UN_corr.shape[1]))
ax.set_yticklabels(df_UN_corr.columns, fontsize=8)  # y axis labels

# Add correlation values as text annotations in each cell
for i in range(corr.shape[0]):
    for j in range(corr.shape[1]):
        text = ax.text(j, i, f'{corr.iloc[i, j]:.2f}',  # Format to 2 decimal places
                       ha="center", va="center", color="black", fontsize=7)  # Text formatting
        
# Add colorbar and title
cb = plt.colorbar(im, ax=ax)  # add a colour legend (called colorbar)
cb.ax.tick_params(labelsize=14)  # add font size
ax.set_title('Correlation Matrix', fontsize=14)  # add title

# Display the plot
plt.show()
```

## The positive numbers mean a positive correlation and the negative numbers mean a negative correlation. 

## Of particular note are:

## the Crude Death Rate (CDR) being positively associated with Meadian Age of the Population (0.5)

## As birth rate increases so does the Mortality-rate under 60 (0.83)

## Mortality rate under 60 decreases with increasing Median Age of Population (-0.79)

## As median Age of a country increases, Crude Birth Rate decreases (-0.91)

## As % of men in a country decrease, crude death rate increases (-0.44)
