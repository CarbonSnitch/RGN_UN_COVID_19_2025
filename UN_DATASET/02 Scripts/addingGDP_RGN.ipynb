```python
path = r"C:\Users\ryani\Desktop\Data Analyst COURSE\Data Immersion\E6 - Dashboards and Analytics\UN_DATASET"

# Adjust the path if needed
gdp_df = pd.read_csv(os.path.join(path, '01 Original Data', 'world_gdp_per_capita_2015_2024.csv'))
```

```python
#all columns
pd.set_option('display.max_columns', None)
```

```python
gdp_long = (
    gdp_df
      .melt(
          id_vars=['country'],
          var_name='Year',
          value_name='GDP_per_capita'
      )
      .dropna(subset=['GDP_per_capita'])  # remove years with missing GDP
)
# Ensure Year is numeric so it matches df_UN
gdp_long['Year'] = gdp_long['Year'].astype(int)
```

```python
df_UN = pd.read_csv(os.path.join(path, '03 Prepared Data', 'UN_df_cleaned.csv'))
df_UN.head()
```

```python
gdp_long.head()
```

```python
gdp_long = gdp_long.rename(columns={'Year': 'Time'})
gdp_long
```

```python
gdp_long = gdp_long.rename(columns={'country': 'Location'})
gdp_long
```

```python
df_merged = df_UN.merge(
    gdp_long,
    left_on=['Location', 'Time'],
    right_on=['Location', 'Time'],
    how='left'
)


df_merged
```

```python
df_USA = df_merged[df_merged['ISO3_code' == 'USA']]
df_USA
```

```python
df_merged.isna().sum()
```

```python
# Group by country and fill NA values with the mean GDP_per_capita for each country
df_merged['GDP_per_capita'] = df_merged.groupby('Location')['GDP_per_capita'].transform(
    lambda x: x.fillna(x.mean())
)

# If there are countries with all NA values in GDP_per_capita, you might want to handle those separately
# This will fill any remaining NAs with the overall mean
remaining_na_mask = df_merged['GDP_per_capita'].isna()
if remaining_na_mask.any():
    overall_mean = df_merged['GDP_per_capita'].mean()
    df_merged.loc[remaining_na_mask, 'GDP_per_capita'] = overall_mean
    
# Check if there are any remaining NAs
print(f"Remaining NAs in GDP_per_capita: {df_merged['GDP_per_capita'].isna().sum()}")
```
